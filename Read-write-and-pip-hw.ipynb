3. Запишите в файл funnel.csv визиты из файла visit_log.csv, в которых были покупки с указанием категории.

Учтите условия на данные:
- содержимое purchase_log.txt помещается в оперативную память компьютера
- содержимое visit_log.csv - нет; используйте только построчную обработку этого файла

````python
import json

purchases = {}
visit_purchase = {}

with open('purchase_log.txt', encoding = 'utf8') as f, open('visit_log.csv', encoding = 'utf8') as f1, open('funnel.csv', 'w' ) as f2:

    f2.write(f'"user_id"; "source"; "category"\n') 
    
    for line in f:
        next(f)
        line_ = json.loads(line)
        key_name = line_.get('user_id')
        purchases.setdefault(key_name, [])
        purchases[key_name].append(line_.get('category'))
        
    for line1 in f1:   
        line1_ = line1.strip().split(',')

        if line1_[0] in purchases.keys():
            type_purchases = ','.join(purchases.get(line1_[0]))
            f2.write(f'{line1_[0]}; {line1_[1]}; {type_purchases}\n') 
````
